### Populating source Logitude and Latitude

```python
for country in data1_2008['Origin Country'].unique():
    if (country == "Sub-Saharan Africa"):
        location = geolocator.geocode("Africa")
    else:
        if (country == "Former Soviet states" or country == "Former Soviet Union" ):
            location = geolocator.geocode("Soviet union")
        else:
            if (country != "Neighboring countries"):
                location = geolocator.geocode(country)
    data1_2008.loc[(data1_2008['Origin Country']==country), 'Source_Latitude'] = location.latitude
    data1_2008.loc[(data1_2008['Origin Country']==country), 'Source_Longitude'] = location.longitude
```

### Populating Destination Logitude and Latitude

```python
for country in data1_2008['Destination Country'].unique():
    if (country == "Sub-Saharan Africa"):
        location = geolocator.geocode("Africa")
    else:
        if (country == "Former Soviet states" or country == "Former Soviet Union" ):
            location = geolocator.geocode("Soviet union")
        else:
            if (country != "Neighboring countries"):
                location = geolocator.geocode(country)
    data1_2008.loc[(data1_2008['Destination Country']==country), 'Destination_Latitude'] = location.latitude
  #  data1_2008.loc[(data1_2008['Destination Country']==country), 'Destination_Longitude'] = location.longitude
```
